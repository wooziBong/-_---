불필요한 데이터 삭제하기

- 열 삭제 -> loc 메서드, drop() 메서드
- 행 삭제 -> []연산자
- 중복 행 찾기 -> group by + sum 메서드 -> update 메서드 -> set_index, reset_index 메서드


```python
# 열 삭제하기
import gdown

# 남산 도서관 데이터 다운로드
gdown.download('https://bit.ly/3RhoNho', 'ns_202104.csv', quiet=False)
```

```python
import pandas as pd

ns_df = pd.read_csv('ns_202104.csv', low_memory=False)
ns_df.head()

#결과의 Nan 값은 누락된 값, 비어 있는 값을 의미한다.
```

```python
# loc 메서드의 슬라이싱 이용
ns_book = ns_df.loc[:, '번호': '등록일자'] # 번호 열부터 등록일자 열까지 전체 행을 선택한다.
ns_book.head()
```

```python
# loc 메서드와 불리언 배열
# 데이터프레임 열 이름이 저장된 columns 속성 확인 -> 판다스의 Index 클래스 객

print(ns_df.columns)
print(ns_df.columns[0])
```

```python
# 원소별 비교를 활용해 Unnamed:13 열이 아닌 것을 표시하는 배욜 생성

ns_df.columns != 'Unnamed:13'
```

```python
# 원소별 비교 연산으로 불리언 배열을 얻고 selected_coulmns 변수에 저장하고 판다스 데이터프레임의 loc 메서드에 전달

selected_columns = ns_df.columns != 'Unnamed: 13'

ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# 부가기호 열 제외하고 선택하기
# 중간에 있는 열 제외 시 슬라이싱 사용이 어려움 -> 불리언 배열 활용 (원하는 열만 선택 가능)

selected_columns = ns_df.columns != '부가기호'

ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# drop 메서드 -> 데이터프레임의 열이나 행을 삭제한다.
# 첫 번째 매개변수 -> 열 이름, axis 매개변수 -> 1로 지정

ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

```python
# drop 메서드로 중간에 있는 열 제외 가능

ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis=1)
ns_book.head()
```

```python
# drop 메서드의 inplace 매개변수 -> 현재 선택한 데이터프레임을 바로 수정할 수 있다.

ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

```python
# dropna() 메서드 -> NaN이 하나 이상 포함된 행이나 열을 삭제한다.

ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
# 모든 값이 NaN인 열 삭제 -> dropna 메서드의 how 매개변수를 all로 지정

ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```

```python
# 행 삭제하기 -> axis 매개변수를 0으로 지정하면 행을 삭제할 수 있다.

ns_book2 = ns_book.drop([0,1])
ns_book2.head()
```

```python
# [] 연산자와 슬라이싱
# [] 연산자 -> 열 이름 또는 열 이름 리스트를 전달하여 데이터프레임의 열을 선택할 수 있다.
# 슬라이싱이나 불리언 배열을 전달하면 행을 선택한다.

# ns_booktes = ns_book[2:] # 인덱스 2부터 출력
# ns_booktes. head()

ns_book2 = ns_book[0:2] # 인덱스 0과 1만 선택, 2 포함 x
ns_book2.head()
```

```python
# [] 연산자와 불리언 배열
# 출판사가 한빛미디어인 행만 선택한다.
selected_rows = ns_df['출판사'] == '한빛미디어'
ns_book2 = ns_book[selected_rows]
ns_book2.head()
```

```python
# loc 메서드에서도 불리언 배열을 사용해 행을 선택할 수 있다.

ns_book2 = ns_book.loc[selected_rows]
ns_book2.head()
```

```python
# 대출건수가 1,000 이하인 행 모두 삭제 or 제외 - 불리언 배
# loc, drop, dropna 메서드 모두 사용 가능 -> 열, 행 삭제

ns_book2 = ns_book[ns_book['대출건수']>1000]
ns_book2.head()
```

```python
# 중복된 행 찾기 - duplicated
# 데이터프레임에 있는 모든 열을 기준으로 중복된 행을 찾는다.

# 해당 명령은 0으로 출력된다. -> 번호 열에 고유한 값을 가지고 있기 때문에 중복된 행이 나올 수 없다.
sum(ns_book.duplicated()) # 중복된 행이 없다면 값이 0으로 출력된다.
```

```python
# duplicated 메소드의 subset 매개변수 활용
sum(ns_book.duplicated(subset=['도서명','저자','ISBN']))
```

```python
# duplicated 메소드의 keep 매개변수 활용

# keep 매개변수 False 지정 -> 중복된 행을 True로 표시한다.
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

```python
# 중복된 행을 합치지 않고 그냥 삭제 -> drop_duplicates 메서드 사용
```

```python
# 그룹별 모으기 - groupby 메서드 - by 매개변수 -> 행을 합칠 때 기준이 되는 열 지정

count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
count_df
```

```python
# sum -> 같은 책의 대출건수는 하나로 합한다.

group_df = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False)
loan_count = group_df.sum()

loan_count
```

```python
loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
loan_count.head()
```

```python
# 원본 데이터 업데이트하기

dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
unique_rows = ~dup_rows # 불리언 배열을 반전 시킨다. ~ 연산자

ns_book3 = ns_book[unique_rows].copy()
```

```python
sum(ns_book3.duplicated(subset=['도서명','저자','ISBN','권']))
```

```python
# set_index -> 지정한 열을 인덱스로 설정한다.

ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
ns_book3.head()
```

```python
# 업데이트하기 update
# 다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트할 때 사용한다.

ns_book3.update(loan_count)
ns_book3.head()
```

```python
# 데이터프레임 인덱스를 재설정한다. reset_index

ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
# 대출건수 확인
sum(ns_book['대출건수']>100)
```

```python
sum(ns_book4['대출건수']>100)
```

```python
# []연산자를 활용해 원하는 열 이름을 순서대로 전달한다. -> 열 순서를 바꾼다.

ns_book4 = ns_book4[ns_book.columns]
ns_book4.head()
```

```python
# 데이터프레임을 csv 파일로 저장한다.

ns_book4.to_csv('ns_book4.csv', index=False)
```
