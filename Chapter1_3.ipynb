open() 함수를 통해 CSV 파일 읽기

readline() 메서드 -> 파일 한 줄씩 읽기

chardet 패키지의 chardet.detect() 함수를 통해 문자 인코딩 방식 파악이 가능하다.

encoding 매개변수를 지정해 파일을 읽는다.


```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2023년 10월).csv', encoding='EUC-KR', low_memory=False)
```

read_csv 함수를 통해 CSV 파일을 읽는다.

low_memory 매개변수 -> False로 지정하면 파일을 나누어 읽지 않고 한 번에 읽는다.
=> 파일을 한 번에 읽기에 메모리 소모가 심하다.

low_memory와 같이 dtype 매개변수를 이용하여 데이터 타입을 지정하는 것도 좋은 방법이다.

```python
df.head()
```

첫 번째 열은 데이터프레임의 인덱스

- CSV 파일의 첫 행이 열 이름이 아니다. -> read_csv() 함수 호출 시 header 매개변수를 None으로 지정한다.
-> name 매개변수에 열 이름 리스트를 따로 전달한다. ( 전달하는 열 이름에 중복된 이름이 있으면 안 된다.)


read_table() 함수는 콤마 대신 탭으로 구분된 파일을 읽는다.


```python
# to_csv() 메서드를 통해 데이터프레임을 csv 파일로 저장한다.
df.to_csv('ns_20231101.csv') # 실행 시 인덱스가 포함되어 저장된다.
df.to_csv('ns_20231101_2.csv', index=False) # 실행 시 인덱스가 포함되지 않는다.

# to_excel() 메서드를 통해 데이터프레임을 엑셀로 저장한다.
df.to_excel('ns_20231101_Excel.xlsx', index=False)
```

```python
with open('ns_20231101.csv') as f:
  for i in range(3):
    print(f.readline(), end='')

# 실행 시 데이터프레임의 행 인덱스가 함께 저장되어 출력된다.
```
